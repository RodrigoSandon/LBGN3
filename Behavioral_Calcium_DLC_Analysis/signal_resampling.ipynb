```python
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
x = np.linspace(0, 10, 20, endpoint=False)
y = np.cos(-x**2/6.0)

f_fft = signal.resample(y, 100)
f_poly = signal.resample_poly(y, 100, 20, padtype='line')  # same as "signal.resample_poly(y, 5, 1, padtype='line')"

xnew = np.linspace(0, 10, 100, endpoint=False)


plt.figure(figsize=(10, 4))
plt.plot(xnew, f_fft, 'b.-', xnew, f_poly, 'r.-')
plt.plot(x, y, 'ko-')
plt.plot(10, y[0], 'bo', 10, 0., 'ro')  # boundaries
plt.legend(['resample', 'resamp_poly', 'data'], loc='best')
plt.tight_layout()
plt.show()
```

Upsampling

```python
x = np.arange(0, len(sig_1))
x_2 = np.arange(0, len(sig_2))
y = sig_1

f_fft = signal.resample(y, len(sig_2))
print(len(f_fft))
f_poly = signal.resample_poly(y, len(sig_2), 1, padtype='line')  # same as "signal.resample_poly(y, 5, 1, padtype='line')"
print(len(f_poly))

xnew = np.arange(0, len(sig_2))


plt.figure(figsize=(10, 4))
plt.plot(xnew, f_fft, 'b-')
plt.plot(x, y, 'k-')
plt.plot(x_2, sig_2, 'r-')
plt.plot(10, y[0], 'bo', 10, 0., 'ro')  # boundaries
plt.legend(['resampled sig 1', 'sig 1', 'sig 2'], loc='best')
plt.tight_layout()
plt.show()
```
